```python
# INSTALAR BIBLIOTECAS
!pip install pandas openpyxl reportlab

import pandas as pd
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas

# =========================
# 1. LER A PLANILHA
# =========================

df = pd.read_csv("Imobacademy.csv", encoding='latin1', delimiter=';')

# =========================
# 2. FILTRAR DADOS
# =========================

# Exemplo: apenas imóveis ativos
df_filtrado = df[df["Status"] == "Ativo"]

# =========================
# 3. CRIAR PDF DAS ETIQUETAS
# =========================

pdf = canvas.Canvas("etiquetas.pdf", pagesize=A4)

# posição inicial
x = 40
y = 800

for index, row in df_filtrado.iterrows():

    texto = f"""
    Imobiliária: {row['Imobiliária']}
    Ticket médio: {row['Ticket médio']}
    Vendas: {row['Vendas']}
    Faturamento: {row['Faturamento']}
    """

    # desenha a etiqueta
    pdf.rect(x, y - 50, 250, 60)

    pdf.drawString(x + 10, y - 15, f"Imobiliária: {row['Imobiliária']}")
    pdf.drawString(x + 10, y - 30, f"Ticket médio: {row['Ticket médio']}")
    pdf.drawString(x + 10, y - 45, f"Vendas: {row['Vendas']}")
    pdf.drawString(x + 10, y - 60, f"Faturamento: {row['Faturamento']}")

    # próxima etiqueta
    y -= 90

    # quebra de página
    if y < 100:
        pdf.showPage()
        y = 800

# =========================
# 4. SALVAR PDF
# =========================

pdf.save()

print("Etiquetas geradas com sucesso!")
```
