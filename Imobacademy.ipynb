```python
# INSTALAR BIBLIOTECAS ESSE
!pip install pandas openpyxl reportlab

import pandas as pd
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
import os

# =========================
# 1. LER PLANILHA
# =========================

arquivo_csv = "Imobacademy.csv"

# verifica se o arquivo existe
if not os.path.exists(arquivo_csv):
    print(f"Arquivo '{arquivo_csv}' nÃ£o encontrado!")
    exit()

df = pd.read_csv(arquivo_csv, encoding='utf-8-sig', delimiter=';')

# =========================
# 2. FILTRAR DADOS
# =========================

df_filtrado = df[df["Status"] == "Ativo"]

# =========================
# 3. GERAR PDF
# =========================

nome_pdf = "Etiquetas para imprimir.pdf"

pdf = canvas.Canvas(nome_pdf, pagesize=A4)

# 'x' e 'y' foram comentados no cÃ³digo original, por isso adicionei-os novamente
x = 40
y = 800

x_positions = [40, 300]  # duas colunas
x_index = 0

for index, row in df_filtrado.iterrows():
    # Atualiza a posiÃ§Ã£o x para a coluna atual
    x = x_positions[x_index]

    pdf.rect(x, y - 65, 150, 70)

    pdf.drawString(x + 10, y - 10, f"ImobiliÃ¡ria: {row['ImobiliÃ¡ria']}")
    pdf.drawString(x + 10, y - 25, f"Vendas: {row['Vendas']}")
    pdf.drawString(x + 10, y - 40, f"Ticket mÃ©dio: {row['Ticket mÃ©dio']}")
    pdf.drawString(x + 10, y - 55, f"Faturamento: {row['Faturamento']}")

    # alterna entre esquerda e direita
    x_index += 1

    # quando completar 2 colunas
    if x_index >= 2:
        x_index = 0
        y -= 110

    # nova pÃ¡gina
    if y < 100:
        pdf.showPage()
        y = 800

# salva o PDF
pdf.save()

# =========================
# 4. CONFIRMAÃÃO
# =========================

print(f"PDF gerado com sucesso: {nome_pdf}")
print("Local salvo:", os.getcwd())
# faz download automÃ¡tico
from google.colab import files

files.download("Etiquetas para imprimir.pdf")
```

```python
# INSTALAR BIBLIOTECAS
!pip install twilio

from flask import Flask, request
from twilio.twiml.messaging_response import MessagingResponse
import pandas as pd
from datetime import datetime

app = Flask(__name__)

# fluxo simples em memÃ³ria
usuarios = {}

@app.route("/bot", methods=["POST"])
def bot():

    numero = request.values.get("From")
    mensagem = request.values.get("Body").strip()

    resposta = MessagingResponse()

    # inicia conversa
    if numero not in usuarios:

        usuarios[numero] = {"etapa": 1}

        resposta.message(
            "OlÃ¡! ð\n\nDeseja preencher seu check-in diÃ¡rio?\n\n1ï¸â£ Sim\n2ï¸â£ NÃ£o"
        )

        return str(resposta)

    etapa = usuarios[numero]["etapa"]

    # etapa 1
    if etapa == 1:

        if mensagem == "1":

            usuarios[numero]["etapa"] = 2

            resposta.message(
                "Quantos imÃ³veis vocÃª captou hoje?"
            )

        else:

            resposta.message("Tudo bem ð")

    # etapa 2
    elif etapa == 2:

        usuarios[numero]["captacoes"] = mensagem
        usuarios[numero]["etapa"] = 3

        resposta.message(
            "Quantos atendimentos realizou hoje?"
        )

    # etapa 3
    elif etapa == 3:

        usuarios[numero]["atendimentos"] = mensagem

        # salva dados
        dados = {
            "Data": [datetime.now()],
            "Telefone": [numero],
            "CaptaÃ§Ãµes": [usuarios[numero]["captacoes"]],
            "Atendimentos": [mensagem]
        }

        df = pd.DataFrame(dados)

        try:
            existente = pd.read_excel("checkins.xlsx")
            df = pd.concat([existente, df], ignore_index=True)
        except:
            pass

        df.to_excel("checkins.xlsx", index=False)

        resposta.message(
            "Check-in concluÃ­do com sucesso â"
        )

        del usuarios[numero]

    return str(resposta)

if __name__ == "__main__":
    app.run()
```
